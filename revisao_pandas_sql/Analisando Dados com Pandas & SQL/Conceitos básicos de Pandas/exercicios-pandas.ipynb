### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df_rj = df_gasolina_comum[df_gasolina_comum['ESTADO'] == 'RIO DE JANEIRO'].copy()
```

```python
#Agrupar por ano e calcular a média de preço
media_anual_rj = df_rj.groupby(df_rj['DATA FINAL'].dt.year)['PREÇO MÉDIO REVENDA'].mean()
```

```python
#3. Calcular a variação percentual entre ano atual e o anterior
variacao_anual = media_anual_rj.pct_change() * 100
```

```python
tabela_variacao = pd.DataFrame({
    'Preço Médio (R$)': media_anual_rj,
    'Variação (%)': variacao_anual
})

display(tabela_variacao.round(2))
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
idx_min = df_gasolina_comum.groupby(df_gasolina_comum['DATA FINAL'].dt.year)['PREÇO MÉDIO REVENDA'].idxmin()
idx_max = df_gasolina_comum.groupby(df_gasolina_comum['DATA FINAL'].dt.year)['PREÇO MÉDIO REVENDA'].idxmax()

df_min = df_gasolina_comum.loc[idx_min, ['DATA FINAL', 'ESTADO', 'PREÇO MÉDIO REVENDA']]
df_max = df_gasolina_comum.loc[idx_max, ['DATA FINAL', 'ESTADO', 'PREÇO MÉDIO REVENDA']]

df_min['Ano'] = df_min['DATA FINAL'].dt.year
df_max['Ano'] = df_max['DATA FINAL'].dt.year

tabela_final = pd.merge(
    df_min[['Ano', 'ESTADO', 'PREÇO MÉDIO REVENDA']], 
    df_max[['Ano', 'ESTADO', 'PREÇO MÉDIO REVENDA']], 
    on='Ano', 
    suffixes=('_MIN', '_MAX')
).set_index('Ano')

tabela_final['DIF_ABSOLUTA'] = tabela_final['PREÇO MÉDIO REVENDA_MAX'] - tabela_final['PREÇO MÉDIO REVENDA_MIN']
tabela_final['DIF_PERCENTUAL'] = (tabela_final['DIF_ABSOLUTA'] / tabela_final['PREÇO MÉDIO REVENDA_MIN']) * 100
```

```python
display(tabela_final.round(2))
```
